#### <center> SKEWNESS(ASIMETRÍA)  </center>


##### Cuando hablamos de skewness, hablamos del concepto de asimetría, una medida de distribución de la probabilidad. </br> En palabras simples, la forma que los datos adquieren al estar distribuidos alrededor de su media.</br></br>Asimetría a la derecha: la curva de la distribución se extiende más hacia la derecha (o tiene un "cola" más larga hacia la derecha) de la media. Este tipo de asimetría se refleja en una medida de asimetría positiva </br></br>Asimetría a la izquierda: la curva de la distribución se extiende más hacia la izquierda (o tiene un "cola" más larga hacia la izquierda) de la media. Este tipo de asimetría se refleja en una medida de asimetría negativa.</br></br>Simetría: la curva de la distribución es simétrica alrededor de la media. Esto se refleja en una medida de asimetría cercana a cero.


#### <center> CONSIDERACIONES  </center>

##### Antes de calcular la asimetría (skewness) de un conjunto de datos, es importante realizar algunas tareas de limpieza y preparación de los datos. Algunas de las tareas más importantes a considerar son: </br></br> **Eliminar valores faltantes:** es importante eliminar los valores faltantes del conjunto de datos antes de calcular la asimetría, ya que los valores faltantes pueden afectar significativamente el resultado.</br></br>**Verificar la normalidad:** es importante verificar si los datos siguen una distribución normal antes de calcular la asimetría, ya que si los datos no siguen una distribución normal, la asimetría puede no ser un indicador útil de la forma de la distribución.</br></br>**Escalar los datos:** es importante escalar los datos antes de calcular la asimetría, ya que esta medida es sensible a las unidades de medida de los datos.</br></br>**Analizar outliers:** es importante analizar los valores atípicos o outliers antes de calcular la asimetría, ya que estos valores pueden afectar significativamente el resultado.</br></br>  **Analizar la relación entre variables:** es importante analizar la relación entre variables antes de calcular la asimetría, ya que esta medida puede ser afectada por la relación entre las variables.

#### <center> VERIFICANDO DISTRIBUCIONES EN NUESTRO CONJUNTO DE DATOS  </center>


```python
from scipy import stats

# Asignamos una variable a analizar, en este caso la variable será el progreso de la enfermedad
y = diabetes_data.target

# Verificamos si el conjunto de datos sigue una distribución normal
_, p_value = stats.shapiro(y)
if p_value > 0.05:
    print("Los datos siguen una distribución normal")
else:
    print("Los datos no siguen una distribución normal")
```

######  La función shapiro() devuelve un valor p, si el valor p es mayor a 0.05 se considera que los datos siguen una distribución normal, en caso contrario no.</br>  Ten en cuenta que este test solo es válido para conjuntos de datos con un tamaño suficientemente grande, para tamaños pequeños se recomienda utilizar otro test de normalidad como el test de kolmogorov-smirnov o el test de Anderson-Darling.

#### <center> EJEMPLO DE ESCALADO DE DATOS  </center>


```python
from sklearn.preprocessing import MinMaxScaler

# Asignamos una variable a analizar, en este caso la variable será el progreso de la enfermedad
y = diabetes_data.target

# Creamos una instancia del escalador
scaler = MinMaxScaler()

# Ajustamos el escalador con los datos
scaler.fit(y.reshape(-1, 1))

# Escalamos los datos
y_scaled = scaler.transform(y.reshape(-1, 1))

# Verificamos el rango de los datos escalados
print("Mínimo:", y_scaled.min())
print("Máximo:", y_scaled.max())
```

```python
#Comprobación gráfica
plt.hist(y_scaled, bins = 30, color = "blue", alpha = 0.5)
plt.xlabel("Valores escalados")
plt.ylabel("Frecuencia")
plt.show()
```

#### <center> ANALIZANDO OUTLIERS  </center>


```python
# Asignamos una variable a analizar, en este caso la variable será el progreso de la enfermedad
y = diabetes_data.target

# Obtenemos los percentiles
q1, q3 = np.percentile(y, [25, 75])

# Intervalo intercuartil
iqr = q3 - q1

# Obtenemos los valores mínimo y máximo permitidos
min_value = q1 - (1.5 * iqr)
max_value = q3 + (1.5 * iqr)

# Filtramos los outliers
y_filtered = y[(y > min_value) & (y < max_value)]
```

```python
# Creamos el gráfico de caja y box
plt.boxplot(y, vert=False)
plt.show()
```

#### <center> EJEMPLO DE REGRESIÓN EN DIABETES  </center>


```python
# Asignamos las variables independientes y la variable dependiente
X = diabetes_data.data
y = diabetes_data.target

# Dividimos los datos en conjunto de entrenamiento y conjunto de prueba
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=33)

# Creamos una instancia del modelo de regresión lineal
reg = LinearRegression()

# Entrenamos el modelo con los datos de entrenamiento
reg.fit(X_train, y_train)

# Hacemos predicciones con el conjunto de prueba
y_pred = reg.predict(X_test)

# Evaluamos el modelo con la métrica R-squared
score = reg.score(X_test, y_test)
print("R-squared: ", score)

#Recuerda :
#Un valor de R-squared cercano a 1 indica un ajuste fuerte y una relación positiva entre las variables independientes y la variable dependiente.
#Un valor de R-squared cercano a 0 indica un ajuste débil y una relación débil o nula entre las variables independientes y la variable dependiente.

```
